# Q 4.Display data by splitting age in 4 quartiles and labeling the quartiles.

```python
#  creating Age colum
df_Patients = df_Patients.copy()
df_Patients['DateOfBirth'] = pd.to_datetime(df_Patients['DateOfBirth'])
df_Patients['Year'] = df_Patients['DateOfBirth'].dt.year
df_Patients['Age'] = 2023 - df_Patients['Year']
df_Patients
```

```python
# Splitting Age Column into Four Quantiles

df_Patients['Age Groups'] = pd.qcut(df_Patients['Age'], 4)
print(df_Patients.head())

# Adding labels to the groupings

df_Patients['Age Group'] = pd.cut(
    df_Patients['Age'], 
    [0, 17, 64, 100], 
    labels=['0-18 years old', '18-65 years old', '65+ years old']
)
print(df_Patients.head())
```

# Q 79. Which PrimaryDiagnosis has the biggest difference between maximum and minimum Expected LOS?

```python
df_Discharges_S = df_Discharges.copy()
ndf = df_Discharges_S.assign(ExpLosMin = df['ExpectedLOS'].abs(),ExpLosMax=df['ExpectedLOS'].abs()).groupby('PrimaryDiagnosis')\
     .agg({'ExpLosMin':'min','ExpLosMax':'max'})
ndf
```

```python
ndf['Max_Min_diff'] = ndf['ExpLosMax'] - ndf['ExpLosMin']
ndf
```

```python
maxDff = ndf['Max_Min_diff'].max()
 
print("Maximum value in column 'Max_Min_diff': " )
print(maxDff)
```

```python
print(" The PrimaryDiagnosis that has the biggest difference between maximum and minimum Expected LOS is : " )
ndf = ndf.loc[(ndf['Max_Min_diff'] == maxDff)]
ndf
```
