```python
df_AmbulatoryVisits_K= df_AmbulatoryVisits.copy()
df_AmbulatoryVisits_K['DateofVisitYear'] = df_AmbulatoryVisits_K['DateofVisit'].dt.year
df_AmbulatoryVisits_K['DateScheduledYear'] = df_AmbulatoryVisits_K['DateScheduled'].dt.year
#currentyear = datetime.now().year
#df_PatientCopy['Age'] = currentyear - df_PatientCopy['BirthYear']
df_AmbulatoryVisits_K
```

```python
# Write any Database Name you like, this will create a .db file in the same folder 
# where you're storing this Jupyter Notebook file
conn = sqlite3.connect('Pings_with_Python_SQL.db')
df_AmbulatoryVisits_K.to_sql('AmbulatoryVisits', conn, if_exists='replace', index=False)
# Create a cursor object
cur = conn.cursor()
select_all = "SELECT DateofVisit, DateofVisitYear, DateScheduled, DateScheduledYear FROM AmbulatoryVisits"

rows = cur.execute(select_all).fetchall()
 
# Output to the console screen
for r in rows:
    print(r)

# Close connection to SQLite database
conn.close()
```
