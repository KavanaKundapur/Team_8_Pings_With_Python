# Q 33. Using a bar chart, which Service had the lowest count of Expected Mortality?

```python
df=pd.read_excel('HospitalDatabase.xlsx', 'Discharges')
datapd=pd.pivot_table(data=df, index='Service', values='ExpectedMortality', aggfunc='mean').reset_index().round(10).sort_values(by="ExpectedMortality")
datapd
```

### Answer:As seen above Cardiology has lowest expected mortality

### Method 1 Using Simple Bar Chart (shows lowest count in yellow color)

```python
# Now plotting bar chart
plt.rcParams["figure.figsize"] = (10,8)
sns.set_style('whitegrid')
x=datapd['Service']
y=datapd['ExpectedMortality']
ax=plt.bar(x,y)

# display lowest expected mortality in yellow color 
patch_h = [patch.get_height() for patch in ax.patches]   
lowest = np.argmin(patch_h)   
ax.patches[lowest].set_facecolor('yellow')

#plot titles
plt.title('ExpectedMortality by Service')
plt.xlabel('Service')
plt.ylabel('Expected Mortality')
plt.show()
```

### Method 2 More Efficient way

```python
#Same can be achieved using Seaborn Bar pLot
#It automatically averages and give the result so we don't need to count mean as required in above method
df=pd.read_excel('HospitalDatabase.xlsx', 'Discharges')
ax=sns.barplot(x = 'Service',
            y = 'ExpectedMortality',
            data = df,capsize=0.1,palette='pastel') 
ax.bar_label(ax.containers[0])
plt.xlabel('Service', fontsize=14)
plt.ylabel('ExpectedMortality', fontsize=14)
plt.title('Expected Mortality by Service', fontsize=15)
# Show the plot
plt.show()
```

```python
# Calculate mortality rate by diagnosis
df=pd.read_csv('HospitalDatabase-Discharges.csv')
Diagnosis= df["PrimaryDiagnosis"].unique()
count=int(len(Diagnosis))
i=0
while (i < count):
    df1=df.loc[(df['DischargeDisposition'] == "Expired") & (df['PrimaryDiagnosis'] == Diagnosis[i])]
    df2=df.loc[(df['PrimaryDiagnosis'] == Diagnosis[i])]
    exp_num = int(len(df1))
    total_num= int(len(df2))
    mortality= exp_num/total_num
    print(" Primary Diagnosis: ",Diagnosis[i],"\n","Mortality: ",mortality,"\n")
    i+=1
```
